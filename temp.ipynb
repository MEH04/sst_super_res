```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import os
import time
```

```python
file_path = 'models/model_testing'
for file in os.listdir(file_path):
    if file.startswith('test'): 
        data = np.transpose(np.loadtxt(os.path.join(file_path,file), delimiter=','))
        model_name = (file.split('_')[-1]).split('.')[0]
        plt.plot(data[0],data[1],label=model_name,linewidth=2)
        plt.ylim(27,29)
        plt.xlabel('Epoch', fontsize=15)
        plt.ylabel('PSNR (dB)', fontsize=15)
        plt.legend(fontsize=14)
plt.grid()
plt.savefig('plots/final_plots/model_comparison.png',dpi=1400)
plt.show()
```

```python
file_path = 'models/lr_search_2'
fig, ax = plt.subplots(1, 2, figsize=(8, 8))

for file in os.listdir(file_path):
    if file.startswith('test'):
        data = np.transpose(np.loadtxt(os.path.join(file_path, file), delimiter=','))
        lr = float((file.split('_')[-1]).split('csv')[0][:-1])
        lr = round(lr, 9)
        lr_label = "{:.1E}".format(lr)  # Format lr as scientific notation with 1 decimal place
        ax[0].plot(data[0], data[1], label=lr_label, linewidth=2)
        ax[0].set_xlabel('Epoch', fontsize=15)
        ax[0].set_ylabel('PSNR (dB)', fontsize=15)
        ax[0].set_ylim(20, 30)
        ax[0].grid()
        ax[1].plot(data[0], data[1], label=lr_label, linewidth=2)
        ax[1].set_xlabel('Epoch', fontsize=15)
        ax[1].set_ylabel('PSNR (dB)', fontsize=15)
        ax[1].set_ylim(28, 29)
        ax[1].set_title('Zoomed',fontsize=13)
        ax[1].grid()

fig.legend(fontsize=12, loc=(0.85, 0.15))
plt.savefig('plots/final_plots/lr_search.png', dpi=1400)
plt.show()
```

```python
file_path = 'models/edsr'
for file in os.listdir(file_path):
    if 'edsr' and 'test' in file:
        data = np.transpose(np.loadtxt(os.path.join(file_path,file), delimiter=','))
        model_name = (file.split('_')[1])
        plt.plot(data[0],data[1],label=model_name,linewidth=3)
        plt.ylim(27,29)
        plt.xlabel('Epoch', fontsize=15)
        plt.ylabel('PSNR (dB)', fontsize=15)
        plt.xlim(0,100)
handles, labels = plt.gca().get_legend_handles_labels()
order = [2,0,1]
plt.legend([handles[idx] for idx in order],[labels[idx] for idx in order],fontsize=14)
plt.grid()
plt.savefig('plots/final_plots/complexity.png',dpi=1400)
plt.show()
```
